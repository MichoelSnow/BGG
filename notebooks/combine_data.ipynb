# Imports

```python
import pandas as pd
import os
```

# Load the data

```python
fl_path = '/home/msnow/git/bgg/data/kaggle/'
fl_list = [x for x in os.listdir(fl_path) if x.endswith('.csv') and x.find('bgg_top')>0]
fl_list.sort()
fl_list
```

# Create DF of just weekly game rankings

```python
fl_num = 0
df_rank = pd.read_csv(os.path.join(fl_path, fl_list[fl_num]), usecols=['name','BoardGameRank'])
df_rank = df_rank.rename(columns = {'name':fl_list[fl_num][:10]})
df_rank.head()
```

```python
for wk in fl_list[1:]:
    df_tmp = pd.read_csv(os.path.join(fl_path, wk), usecols=['name','BoardGameRank'])
    df_tmp = df_tmp.rename(columns = {'name':wk[:10]})
    df_rank = df_rank.merge(right=df_tmp, how='left', on='BoardGameRank')
```

```python
df_rank
```

```python
df_rank.to_csv(os.path.join(fl_path,'curated','weekly_rankings.csv'), index=False)
```

# Create game lookup DF

```python
static_cols = ['boardgameartist', 'boardgamecategory', 'boardgamedesigner', 'boardgamefamily', 'boardgamemechanic', 'boardgamepublisher', 'id', 'maxplayers', 'minage', 'minplayers', 'name', 'yearpublished']
static_cols2 = ['boardgameartist', 'boardgamecategory', 'boardgamedesigner', 'boardgamefamily', 'boardgamemechanic', 'boardgamepublisher', 'game_id', 'maxplayers', 'minage', 'minplayers', 'name', 'yearpublished']
```

```python
df_lookup = pd.read_csv(os.path.join(fl_path, fl_list[0]), usecols=static_cols)
for wk in fl_list[1:]:
    try:
        df_tmp = pd.read_csv(os.path.join(fl_path, wk), usecols=static_cols)
    except:
        df_tmp = pd.read_csv(os.path.join(fl_path, wk), usecols=static_cols2)
        df_tmp = df_tmp.rename(columns = {'game_id':'id'})
        pass
    df_lookup = pd.concat([df_lookup, df_tmp], ignore_index=True).drop_duplicates('id')
df_lookup.shape
```

```python
df_lookup.to_csv(os.path.join(fl_path,'curated','game_lookup.csv'), index=False)
```

```python
col_to_use = 'yearpublished'
df_pl_time = pd.read_csv(os.path.join(fl_path, fl_list[0]), usecols=['name',col_to_use])
df_pl_time = df_pl_time[['name', col_to_use]]
df_pl_time = df_pl_time.rename(columns = {col_to_use:fl_list[0][:10]})
for wk in fl_list[1:]:
    df_tmp = pd.read_csv(os.path.join(fl_path, wk), usecols=['name',col_to_use])
    df_tmp = df_tmp.loc[df_tmp.name.isin(df_pl_time.name)]
    df_tmp = df_tmp.rename(columns = {col_to_use:wk[:10]})
    df_pl_time = df_pl_time.merge(right=df_tmp, how='inner', on='name')
    df_pl_time = df_pl_time.drop_duplicates(subset='name')
# df_pl_time.shape  
df_pl_time.loc[df_pl_time.var(axis=1, numeric_only=True)>0,:].var(axis=1, numeric_only=True)
```
